# Extracting windows of Transition class

```python
# Creating a list of indices where FoG begins
transition = list()

for i,v in enumerate(list_2s):
    if(i==0):
        transition.append(v)
    else:
        # Picking the indices from where FoG begins by comparing it with the previous value  
        if((list_2s[i]-1) != list_2s[i-1]):
            transition.append(v)
transition
```

```python
# Going back 50 instances for extracting transition windows
for i,v in enumerate(transition):
    transition[i] = v-50
transition
```

```python
# Creating a list of all the indices in the dataframe from where walking occurences continue  
start_1s = list()

for i,v in enumerate(list_1s):
    if(i==0):
        start_1s.append(v)
    else:
        # Adding the index to list from where walking class begins
        if((list_1s[i]-1) != list_1s[i-1]):
            start_1s.append(v)
start_1s
```

```python
# Creating a list of all the indices in the dataframe where walking class occurence ends and FoG starts from the next index
end_1s = list()

for i,v in enumerate(list_1s):
    if(i==1030049):
        end_1s.append(v)
    else:
        # Adding the index to list from where walking class occurence ends
        if((list_1s[i]+1) != list_1s[i+1]):
            end_1s.append(v)
end_1s
```

```python
# Creating a list of dataframes to store all the windows of walking class
# Every instance of the list is one window with 50 readings
import random
walking_dataframe_list = list()
for i in range(238):
    v = start_1s[i]
    j = end_1s[i]
    # Generating a random value from the continuos range of the walking class
    rd = random.randint(v,j-50) 
    # j-50 because if the random value generator picks j then extracting 50 dataframes from there would include FoG occurences as well
    walking_dataframe = pd.DataFrame()
    # Creating a dataframe from the randomly generated value and taking 50 continuous readings as a window
    walking_dataframe = walking_dataframe.append(dataframe_all.iloc[rd:rd+50])
    walking_dataframe = walking_dataframe.reset_index(drop = True)
    # Adding the currently created dataframe to the list of dataframes
    walking_dataframe_list.append(walking_dataframe)
walking_dataframe_list    
```
